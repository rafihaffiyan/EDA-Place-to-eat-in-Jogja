```python
import pandas as pd
 
file =('places-to-eat-in-the-jogja-region/\
places_to_eat_in_the_jogja_region.csv')

data = pd.read_csv(file)
data.head()
```

```python
print("Jumlah duplikasi: ",data.duplicated().sum())
data.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Ambil kolom "Preferensi Makanan" dari dataset
preferensi_makanan = data['Preferensi Makanan']

# Hitung frekuensi preferensi makanan
frekuensi_preferensi = preferensi_makanan.value_counts()

# Buat plot bar untuk menampilkan preferensi makanan
plt.figure(figsize=(10, 6))
frekuensi_preferensi.plot(kind='bar')
plt.title('Preferensi Makanan')
plt.xlabel('Preferensi')
plt.xticks(rotation=0)
plt.ylabel('Jumlah')
plt.show()
```

```python
jenis_suasana = data['Jenis Suasana']

# Hitung frekuensi jenis suasana
frekuensi_jenis_suasana = jenis_suasana.value_counts()

# Buat plot bar untuk menampilkan jenis suasana
plt.figure(figsize=(10, 6))
frekuensi_jenis_suasana.plot(kind='bar')
plt.title('Jenis Suasana')
plt.xlabel('Suasana')
plt.xticks(rotation=0)
plt.ylabel('Jumlah')
plt.show()
```

```python
# Ambil kolom "All You Can Eat atau Ala Carte" dari dataset
konsep_makanan = data['All You Can Eat atau Ala Carte']

# Hitung jumlah toko yang menyediakan konsep All You Can Eat dan Ala Carte
jumlah_all_you_can_eat = konsep_makanan[konsep_makanan == 'All You Can Eat'].count()
jumlah_ala_carte = konsep_makanan[konsep_makanan == 'Ala Carte'].count()

# Buat data untuk pie chart
labels = ['All You Can Eat', 'Ala Carte']
sizes = [jumlah_all_you_can_eat, jumlah_ala_carte]

# Buat pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Persentase Konsep Makanan')
plt.axis('equal')  # Membuat pie chart menjadi lingkaran

# Tampilkan pie chart
plt.show()
```

```python
# Filter data restoran yang memiliki konsep "All You Can Eat"
resto_all_you_can_eat = data[data['All You Can Eat atau Ala Carte'] == 'All You Can Eat']

# Tampilkan tabel restoran All You Can Eat
resto_all_you_can_eat[['Nama Restoran', 'Preferensi Makanan', 'Lokasi Restoran', 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko']]
```

```python
# Filter data restoran yang menyediakan pelayanan khusus pasangan
resto_pelayanan_khusus_pasangan = data[data['Pelayanan Khusus Pasangan (Ya/Tidak)'] == 'Ya']

# Tampilkan data pelayanan khusus pasangan
resto_pelayanan_khusus_pasangan[['Nama Restoran', 'Preferensi Makanan', 'Lokasi Restoran', 'Pelayanan Khusus Pasangan (Ya/Tidak)', 'Harga Rata-Rata Makanan di Toko (Rp)', 'Rating Toko']]
```

```python
# Set opsi untuk menampilkan semua kolom pada output
pd.set_option('display.max_columns', None)

# Filter data berdasarkan kolom "Toko Sering Diskon (Ya/Tidak)" yang bernilai "Ya" dan rating di atas 4.5
restoran_diskon_rating = data[(data['Toko Sering Diskon (Ya/Tidak)'] == 'Ya') & (data['Rating Toko'] > 4.5)]

# Urutkan data berdasarkan rating secara menurun
restoran_diskon_rating = restoran_diskon_rating.sort_values(by='Rating Toko', ascending=False)

# Tampilkan seluruh data pada tabel restoran yang sering diskon dengan rating di atas 4.5
restoran_diskon_rating
```

```python
# Set opsi untuk menampilkan semua kolom pada output
pd.set_option('display.max_columns', None)

# Filter data berdasarkan kolom "Harga Rata-Rata Makanan di Toko (Rp)" yang di bawah 25000 dan rating di atas 4.5
restoran_harga_rating = data[(data['Harga Rata-Rata Makanan di Toko (Rp)'] < 25000) & (data['Rating Toko'] > 4.5)]

# Tampilkan seluruh data pada tabel restoran dengan harga di bawah 25000 dan rating di atas 4.5
restoran_harga_rating
```

```python
# Set opsi untuk menampilkan semua kolom pada output
pd.set_option('display.max_columns', None)

# Filter data berdasarkan kolom "Variasi Makanan" yang mengandung kata "coffee"
restoran_coffee = data[data['Variasi Makanan'].str.contains('Coffee')]

# Tampilkan seluruh data pada tabel restoran dengan variasi makanan yang mengandung kata "coffee"
restoran_coffee
```
